```python
import numpy as np
import matplotlib.pyplot as plt
```

image.png

Lingo: $a_1, a_2, ..., a_n$ are called **logits**.

First, we create a little function to calculate the probabilities from the logits. Note how we divide the logits by the temperature parameter in the function below.

```python
def softmax(logits, temperature=1.0):
    """Compute the softmax probabilities with a temperature parameter."""
    logits = np.array(logits)
    exp_logits = np.exp(logits / temperature)
    return exp_logits / np.sum(exp_logits)

```

Next, a function that creates a bar chart of the probabilities for 5 tokens.

```python
def plot_softmax(temperature):
    """Plot the softmax distribution for a given temperature."""
    logits = np.array([2, 1, 0, -1, -2])  # Example logits
    probabilities = softmax(logits, temperature)

    plt.figure(figsize=(8, 5))
    plt.bar(range(len(logits)), probabilities, color='blue', alpha=0.7)
    plt.xticks(range(len(logits)), labels=[f"Token {i+1}" for i in range(len(logits))])
    plt.xlabel("Tokens")
    plt.ylabel("Probability")
    plt.title(f"Softmax Distribution (Temperature={temperature:.2f})")
    plt.ylim(0, 1)
    plt.grid(axis='y', linestyle='--', alpha=0.6)
    plt.show()

```

```python
plot_softmax(temperature=1.0)
```

This is so meh. Let's make it interactive!

```python
import ipywidgets as widgets
from IPython.display import display
```

```python
# Create an interactive widget
temperature_slider = widgets.FloatSlider(
    value=1.0,
    min=0.1,
    max=2.0,
    step=0.1,
    description='Temperature',
    continuous_update=True
)

# Display interactive plot
widgets.interactive(plot_softmax, temperature=temperature_slider)
```
